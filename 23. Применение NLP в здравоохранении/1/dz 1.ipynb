```python
import re
import shutil
```

```python
shutil.copyfile('./dates.txt', './dates1.txt')
path_to_dates_txt_file = 'dates1.txt'
```

```python
month_dict = {'jan': 1,
              'feb': 2,
              'marc': 3,
              'apr': 4,
              'may': 5,
              'jun': 6,
              'jul': 7,
              'aug': 8,
              'sep': 9, 
              'oct': 10,
              'nov': 11,
              'dec': 12
             }
```

```python
with open(path_to_dates_txt_file) as f:
    lines = f.readlines()
    
with open(path_to_dates_txt_file, "w") as f:
    for line in lines:
        line = line.lower()
        line = re.sub('[:|"|?|()|#|~|[]]', '', line)
        for word, num in month_dict.items():
            line = line.replace(str(word), str(num))
        line = re.sub('[a-z]', '', line)
        f.write(line)
```

```python
reg_1 = "[0-9]{0,2}/[0-9]{1,2}/[0-9]{2,4}"
reg_2 = "[0-9]{1}/ /[0-9]{2,4}"
extracted_contents = {}
for line in lines:
    line_content = line.split('\t')
    extracted_contents[line_content[0]] = re.findall(reg_1, line_content[1])
    
for str_num, date in extracted_contents.items():
    if date == [] is True:
        extracted_contents[line_content[str_num]] = re.findall(reg_2, line_content[str_num+1])
```

```python
your_surname = 'Tychina'
```

```python
for str_num, date in extracted_contents.items():
    for el in date:
        m, d, y = el.split('/')
        if len(m) == 1:
            m = f'0{m}'
        if len(d) == 1:
            d = f'0{d}'
        if len(y) == 2:
            y = f'19{y}'
        new_date = f"{y}-{m}-{d}"
        extracted_contents.update({str_num: new_date})
```
